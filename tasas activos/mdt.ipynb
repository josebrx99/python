```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

ax = tasascap
ax = ax.groupby(["anomes"]).agg({"monto":"sum", "num_obl":"sum", "tasa":["mean", "median", "min", "max", "std"]}).reset_index()#.sort_values("num_obl", ascending=False)#.head(30)
ax.columns = ["mes", "monto", "num_obl", "tasa_mean", "tasa_median", "tasa_min", "tasa_max", "tasa_std"]
ax["tasa_cv"] = ax["tasa_std"]/ax["tasa_mean"].mean()*100
ax["%monto"] = ax["monto"]/ax["monto"].sum()*100
ax["%num_obl"] = ax["num_obl"]/ax["num_obl"].sum()*100
ax["prom_monto"] = ax["monto"]/ax["num_obl"]
#ax.sort_values("monto", ascending=False).head(40)
ax

def plot_time_series(df, date_col="date", value_col="value", title="Serie temporal"):
    """
    Grafica una serie temporal con puntos y líneas que los unen.
    Parámetros:
      - df: DataFrame con las columnas de fecha y valor
      - date_col: nombre de la columna de fecha
      - value_col: nombre de la columna de valor
      - title: título del gráfico
    """
    df = df.copy()
    df[date_col] = pd.to_datetime(df[date_col])

    fig, axy = plt.subplots(figsize=(13, 5))

    # Línea + puntos
    axy.plot(df[date_col], df[value_col], marker='o', linewidth=1.5, markersize=6, label = "Media")
    axy.plot(df[date_col], ax["tasa_median"], marker='o', linewidth=1.5, markersize=6, label = "Mediana")

    # Formato de fechas en eje X
    axy.xaxis.set_major_locator(mdates.AutoDateLocator())
    axy.xaxis.set_major_formatter(mdates.ConciseDateFormatter(axy.xaxis.get_major_locator()))

    axy.set_title(title)
    axy.set_xlabel("Fecha")
    axy.set_ylabel(value_col)
    axy.grid(alpha=0.3)
    axy.legend()
    plt.tight_layout()
    plt.show()

# Llamada
plot_time_series(ax, "mes", "tasa_mean", title="Comportamiento de la tasa de desembolso")
```

```python
def scatter_line(df, yname, xnames):
    for x in xnames:
        plt.figure(figsize=(9,5))

        # Scatter
        plt.scatter(df[x], df[yname], s=20)

        # Correlación
        corr = df[[yname, x]].corr().iloc[0, 1]
        plt.text(
            0.915, 0.04,
            f'r: {corr:.4f}',
            transform=plt.gca().transAxes,
            fontsize=9,
            bbox=dict(facecolor='white', alpha=0.5)
        )

        # Línea de regresión
        ax = df[[x, yname]].dropna()
        m, b = np.polyfit(ax[x], ax[yname], 1)  
        plt.plot(ax[x], m*ax[x] + b, linewidth=2, color = "black")

        # Etiquetas
        plt.xlabel(x.title().replace("_"," "))
        plt.ylabel(yname.title().replace("_"," "))
        plt.title(f"{yname.title().replace("_"," ")} vs {x.title().replace("_"," ")}")
        plt.grid(True, alpha=0.3)
        plt.legend()
        #plt.ylim([df[yname].min()*0.05, df[yname].max()*1.1])
        plt.tight_layout()
        plt.show()

xnames = cors.reset_index()["index"].tolist()[1:5] + cors.reset_index()["index"].tolist()[-5:]
scatter_line(df, yname, xnames)
```
